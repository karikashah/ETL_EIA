```python
# get EIA data using API key, returns a json object 
# print(EIA_APP_KEY)
def state_id_to_json(series_id):
    params = {
        "series_id": series_id, 
        "api_key": EIA_APP_KEY
    }

    target_url = ('http://api.eia.gov/series/')
    response = requests.get(target_url, params=params)
    electricity_results = response.json()
    return electricity_results
```

```python
def insert_into_mongo(state_abbr, annual_wind, annual_hyc, annual_tsn, annual_cow, annual_nuc, annual_ng):
#     print(f"state abbr - {state_abbr}")
    if(annual_tsn): 
        # add solar data (since solar API has all states hence using insert for solar fuel type)
        headers = ['Year','Solar (in thousand megawatthours)']
        df_t = pd.DataFrame(annual_tsn['series'][0]['data'], columns=headers)
        df_t['State'] = state_abbr 
        statewise_collection.insert_many(df_t.to_dict('records'))
    
    if(annual_hyc): 
        # add Hydroelectric data
        headers = ['Year','Hydroelectric (in thousand megawatthours)']
        df_h = pd.DataFrame(annual_hyc['series'][0]['data'], columns=headers)
        df_h['State'] = state_abbr 

        for index, row in df_h.iterrows():
            statewise_collection.update_many({'Year': row[0], 'State': state_abbr}, {"$set": row.to_dict()}, upsert=True)
    
    if(annual_wind): 
        ## - Adding wind column
        headers = ['Year','Wind (in thousand megawatthours)']
        df_w = pd.DataFrame(annual_wind['series'][0]['data'], columns=headers)
        df_w['State'] = state_abbr 

        for index, row in df_w.iterrows():
            statewise_collection.update_many({'Year': row[0], 'State': state_abbr}, {"$set": row.to_dict()}, upsert=True)
    
    if(annual_cow): 
        ## - Adding Coal column
        headers = ['Year','Coal (in thousand megawatthours)']
        df_c = pd.DataFrame(annual_cow['series'][0]['data'], columns=headers)
        df_c['State'] = state_abbr 

        for index, row in df_c.iterrows():
            statewise_collection.update_many({'Year': row[0], 'State': state_abbr}, {"$set": row.to_dict()}, upsert=True)
    
    if(annual_ng): 
        ## - Adding Natural Gas column
        headers = ['Year','Natural Gas (in thousand megawatthours)']
        df_ng = pd.DataFrame(annual_ng['series'][0]['data'], columns=headers)
        df_ng['State'] = state_abbr 

        for index, row in df_ng.iterrows():
            statewise_collection.update_many({'Year': row[0], 'State': state_abbr}, {"$set": row.to_dict()}, upsert=True)

    if(annual_nuc):
        ## - Adding Nuclear column
        headers = ['Year','Nuclear (in thousand megawatthours)']
        df_nuc = pd.DataFrame(annual_nuc['series'][0]['data'], columns=headers)
        df_nuc['State'] = state_abbr 

        for index, row in df_nuc.iterrows():
            statewise_collection.update_many({'Year': row[0], 'State': state_abbr}, {"$set": row.to_dict()}, upsert=True)
```
