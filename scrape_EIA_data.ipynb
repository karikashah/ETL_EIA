```python
# get EIA data using API key, returns a json object 
# print(EIA_APP_KEY)
def state_id_to_json(series_id):
    params = {
        "series_id": series_id, 
        "api_key": EIA_APP_KEY
    }

    target_url = ('http://api.eia.gov/series/')
    response = requests.get(target_url, params=params)
    electricity_results = response.json()
    return electricity_results
```
